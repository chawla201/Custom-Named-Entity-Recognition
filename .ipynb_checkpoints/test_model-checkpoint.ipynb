```python
import os
import csv
import json
import re
import pandas as pd
from pathlib import Path
import spacy
```

## Loading the trained model

```python
nlp = spacy.load("/content/drive/MyDrive/qubitrics_internship_assignment/model")
```

##  Extracting Reciept text from test data

```python
test_text_files = os.listdir("/qubitrics/test_text")
```

```python
len(test_text_files)
```

```python
test_data = pd.DataFrame(columns=["filename", "text"])

test_data["filename"] = test_text_files

data_text = []
for file in test_data["filename"]:
    rec_text = []
    pattern = r"\d+,\d+,\d+,\d+,\d+,\d+,\d+,\d+,(.+)"
    try:
      with open(f"/qubitrics/test_text/{file}") as f:
          lines = f.readlines()
          for line in lines:
              rec_text += re.findall(pattern, line)
    except:
      pass
    data_text.append(" ".join([x.strip() for x in rec_text]))
test_data["text"] = data_text
```

```python
test_data
```

## Creating

```python
train = os.listdir("/qubitrics/train")
```

```python
ent_list = []
for file in train:
    with open(f"/qubitrics/train/{file}") as f:
        entity_dict = json.load(f)
        ent_list.append(entity_dict)
```

```python
memory_dictionary = {"company":[], "address":[], "date":[], "total":[]}
for dictionary in ent_list:
    for key, value in dictionary.items():
        memory_dictionary[key].append(value)
```

```python
for index, row in test_data.iterrows():
  op_dict = {"company": "", "date": "", "address": "", "total": ""}
  doc = nlp(row["text"])
  for ent in doc.ents:
    op_dict[ent.label_] = ent.text
  
  for tag, tag_memory in memory_dictionary.items():
    for tag_value in tag_memory:
        if tag=="total":
            pass   
        elif (re.search(tag_value, row["text"]) != None) and (op_dict[tag] == ""):
            op_dict[tag]=tag_value
            
  print("Entities", op_dict)
  json_object = json.dumps(op_dict, indent = 4)
  with open(f"/content/drive/MyDrive/qubitrics_internship_assignment/output/{row['filename']}", "w") as op:
      op.write(json_object)
```
